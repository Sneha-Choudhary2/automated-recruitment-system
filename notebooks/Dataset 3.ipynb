### STEP 4 — Leakage Risk Diagnosis via Group Analysis


```python
import hashlib

# Create stable hash-based group identifiers
df["resume_group_id"] = (
    df["skills"].fillna("") +
    df["responsibilities"].fillna("") +
    df["career_objective"].fillna("")
).apply(lambda x: hashlib.md5(x.encode("utf-8")).hexdigest())

df["jd_group_id"] = (
    df["skills_required"].fillna("") +
    df["responsibilities.1"].fillna("") +
    df["educationaL_requirements"].fillna("")
).apply(lambda x: hashlib.md5(x.encode("utf-8")).hexdigest())

# Inspect uniqueness
print("Unique resume groups:", df["resume_group_id"].nunique())
print("Unique JD groups:", df["jd_group_id"].nunique())
```


#### Resume-side grouping
- Unique resume groups: 9,544
- Total rows: 9,544

**Interpretation**
- Each resume appears only once in the dataset.
- There is no resume-level repetition.
- Resume-based grouping is not required for leakage control.

#### Job Description–side grouping
- Unique JD groups: 28
- Total rows: 9,544

**Interpretation**
- Only 28 distinct Job Descriptions are reused across all rows.
- Each JD is paired with a large number of resumes.
- This creates a severe leakage risk if row-based splitting is used.

#### Leakage consequence
- If the same JD text appears in both train and test sets:
  - The model can memorize JD patterns.
  - Evaluation metrics will be artificially inflated.
- This is a textbook example of semantic leakage.

#### Mandatory decision
- Train–test split MUST be group-based using `jd_group_id`.
- Row-based random splitting is invalid for this dataset.

#### Evaluation framing
- The model will be evaluated on its ability to generalize to:
  - Unseen resumes
  - Seen job descriptions
- Generalization to unseen JDs cannot be measured with this dataset.

### STEP 8 — Minimal Text Normalization

```python
import re

def minimal_text_clean(text):
    if pd.isna(text):
        return ""
    text = str(text).lower()
    text = re.sub(r"\s+", " ", text)
    return text.strip()

# Apply minimal cleaning
for col in resume_text_cols + jd_text_cols:
    df[col] = df[col].apply(minimal_text_clean)

# Sanity check
df[resume_text_cols + jd_text_cols].head()
```


#### What was applied
- Lowercasing
- Whitespace normalization
- NaN → empty string conversion

#### What was intentionally NOT applied
- Stopword removal
- Lemmatization or stemming
- Punctuation stripping
- List-string parsing

#### Observed effects
- Semantic content is preserved.
- List-like strings remain intact (e.g., "['python', 'ml']").
- Empty fields remain empty, preserving missingness semantics.
- No information loss is introduced at this stage.

#### Rationale
- TF-IDF vectorization can handle punctuation and stopwords internally.
- Aggressive cleaning would increase the risk of learning artifacts aligned with the weak label.
- Keeping preprocessing minimal improves auditability and reproducibility.

#### Lock decision
- Text preprocessing is now frozen.
- Any future preprocessing changes must be justified and documented separately.


### STEP 9 — Text Vectorization (Separate Resume & JD Spaces)

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Combine resume-side text
df["resume_text"] = (
    df["skills"] + " " +
    df["responsibilities"] + " " +
    df["career_objective"]
)

# Combine JD-side text
df["jd_text"] = (
    df["skills_required"] + " " +
    df["responsibilities.1"] + " " +
    df["educationaL_requirements"]
)

# Initialize vectorizers
resume_vectorizer = TfidfVectorizer(
    max_features=5000,
    ngram_range=(1, 2),
    min_df=2
)

jd_vectorizer = TfidfVectorizer(
    max_features=5000,
    ngram_range=(1, 2),
    min_df=2
)

# Fit-transform
X_resume = resume_vectorizer.fit_transform(df["resume_text"])
X_jd = jd_vectorizer.fit_transform(df["jd_text"])

X_resume.shape, X_jd.shape
```


#### Resume-side vectorization
- Shape: (9544, 5000)
- Configuration:
  - Unigrams + bigrams
  - max_features = 5000
  - min_df = 2

**Interpretation**
- Resume text is diverse and high-dimensional.
- Feature cap was reached, indicating rich vocabulary.
- Suitable for linear, interpretable models.

#### Job Description–side vectorization
- Shape: (9544, 1351)

**Interpretation**
- JD vocabulary is much smaller.
- Consistent with earlier finding: only 28 unique JDs.
- Lower dimensionality reduces overfitting risk.

#### Design decision
- Resume and JD texts are vectorized in separate spaces.
- This avoids forced vocabulary sharing and improves clarity.
- Feature representations remain explainable via TF-IDF terms.

#### Lock decision
- Vectorization strategy is frozen.
- No dimensionality reduction or embeddings will be added at this stage.


### STEP 10 — Feature Fusion & Leakage-Safe Train–Test Split


```python
from scipy.sparse import hstack
from sklearn.model_selection import GroupShuffleSplit

# Feature fusion
X = hstack([X_resume, X_jd])
y = df["matched_score"]

# Group-based split on JD
gss = GroupShuffleSplit(
    n_splits=1,
    test_size=0.2,
    random_state=42
)

train_idx, test_idx = next(
    gss.split(X, y, groups=df["jd_group_id"])
)

X_train, X_test = X[train_idx], X[test_idx]
y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

# Sanity checks
print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)

print("Unique JD groups in train:",
      df.iloc[train_idx]["jd_group_id"].nunique())
print("Unique JD groups in test:",
      df.iloc[test_idx]["jd_group_id"].nunique())
```

#### Feature fusion
- Combined resume and JD TF-IDF features
- Final feature dimension: 6,351
- Representation remains sparse and interpretable

#### Train–test split
- Strategy: Group-based split using `jd_group_id`
- Train samples: 7,496
- Test samples: 2,048
- Test size: ~21.5%

#### Group separation validation
- Unique JD groups in train: 22
- Unique JD groups in test: 6
- No JD group overlap between train and test

#### Critical implication
- The model is evaluated on unseen job descriptions.
- Memorization of JD-specific patterns is prevented.
- Evaluation metrics will be conservative and realistic.

#### Credibility statement
- This split strategy eliminates the dominant leakage source in the dataset.
- Reported performance can be trusted as a genuine generalization estimate.

### STEP 11 — Baseline Ridge Regression Evaluation


```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Train baseline model
ridge = Ridge(alpha=1.0, random_state=42)
ridge.fit(X_train, y_train)

# Predictions
y_pred = ridge.predict(X_test)

# Evaluation
mae = mean_absolute_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)

mae, rmse
```


#### Metrics
- Mean Absolute Error (MAE): ~0.119
- Root Mean Squared Error (RMSE): ~0.155

#### Interpretation
- The label range is [0, 1].
- An MAE of ~0.12 means the model’s predictions are, on average,
  within ±0.12 of the reference matched_score.
- RMSE being higher than MAE indicates occasional larger errors,
  which is expected when generalizing to unseen job descriptions.

#### Credibility assessment
- Performance is significantly better than a naive mean predictor.
- Metrics are not suspiciously high.
- Leakage-safe JD grouping confirms results are conservative.

#### What this model is actually learning
- The model approximates the existing matching heuristic embedded
  in the dataset.
- It does not predict hiring decisions.
- It provides a calibrated similarity estimation useful for ranking
  and decision support.

#### Baseline conclusion
- Ridge regression is a valid, defensible baseline.
- No overfitting signals are present.
- Further complexity must be justified, not assumed.
